```python
import numpy as np
import matplotlib.pyplot as plt

marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
students = list(marks.keys())
grades = list(marks.values())

plt.pie(grades, labels = students)
plt.legend(title = 'Students', bbox_to_anchor = (1.05, 1.0), loc = 'upper left')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
y1 = np.array([1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])

x2 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
y2 = np.array([5, 5.5, 6, 6.5, 7, 7.5, 8, 8.5, 9])
              
x3 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4])
y3 = np.array([9, 9.5, 10, 10.5, 11, 11.5, 12, 12.5, 13])

plt.subplot(2, 3, 1)
plt.plot(x1, y1)
plt.plot(x2, y2, color = 'orange')
plt.plot(x3, y3, color = 'green')
plt.title('Lines')

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 6, 7, 8, 9, 10])

plt.subplot(2, 3, 2)
plt.bar(x,y)
plt.title('Bar')

x = np.array([1, 2, 3, 4, 5, 6])

plt.subplot(2, 3, 3)
plt.pie(x)
plt.title('Pie')

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 6, 7, 8, 9, 10])

plt.subplot(2, 3, 4)
plt.plot(x,y)
plt.grid()
plt.title('Grid')

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([5, 6, 7, 8, 9, 10])

plt.subplot(2, 3, 5)
plt.scatter(x,y)
plt.title('Scatter')

x = np.array([1, 2, 3, 4, 5, 6])

plt.subplot(2, 3, 6)
plt.hist(x)
plt.title('Histogram')
```
